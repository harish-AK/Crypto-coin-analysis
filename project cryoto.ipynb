```python
eth['Cumulative Return'] = (1 + eth['returns']).cumprod()
```

```python
eth.head()
```

```python
yfi['Cumulative Return'] = (1 + yfi['returns']).cumprod()
btcb['Cumulative Return'] = (1 + btcb['returns']).cumprod()
```

```python
eth['Cumulative Return'].plot(label='etherium',figsize=(16,8),title='Cumulative Return')
yfi['Cumulative Return'].plot(label='YFI')
btcb['Cumulative Return'].plot(label='BTCB')
plt.legend()
```

```python
from mpl_finance import candlestick_ohlc
from matplotlib.dates import DateFormatter, date2num, WeekdayLocator, DayLocator, MONDAY
# Rest the index to get a column of January Dates
eth_reset = eth.loc['2012-01':'2012-01'].reset_index()
# Create a new column of numerical "date" values for matplotlib to use
eth_reset['date_ax'] = eth_reset['Date'].apply(lambda date: date2num(date))
eth_values = [tuple(vals) for vals in eth_reset[['date_ax', 'Open', 'High', 'Low', 'Close']].values]
mondays = WeekdayLocator(MONDAY)        # major ticks on the mondays
alldays = DayLocator()              # minor ticks on the days
weekFormatter = DateFormatter('%b %d')  # e.g., Jan 12
dayFormatter = DateFormatter('%d')      # e.g., 12
#Plot it
fig, ax = plt.subplots()
fig.subplots_adjust(bottom=0.2)
ax.xaxis.set_major_locator(mondays)
ax.xaxis.set_minor_locator(alldays)
ax.xaxis.set_major_formatter(weekFormatter)
candlestick_ohlc(ax, eth_values, width=0.6, colorup='g',colordown='r');
```
